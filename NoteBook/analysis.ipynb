This is my first analysis project, today i'm going to work on a student score dataset that i have downloaded from the kaggle and find the variables that are influencing the student score

```python
import pandas as pd
df=pd.read_excel("../Data/Maths.csv")
df.head()
```

```python
df.info()
#from this command its confirmed that there's no null value in out dataset
```

```python
df.describe()
#It gives the rough extimation about the data like, minimum and maximum value 
```

```python
import matplotlib.pyplot as plt
df['G3'].hist(bins=10, edgecolor='black')
plt.title('Histogram of G3')
plt.xlabel('Vlaues')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

From the above chart its shows that most of the student have scored an average marks from 10.0 to 12.5 with few scoring extreme values like 0 and 20

```python
import matplotlib.pyplot as plt
df['Studytime'].value_counts().sort_values(ascending=False).plot(kind='bar')
plt.xlabel("Study Time")
plt.ylabel("Number of Students")
plt.title("Study Time Distribution")
plt.show()
```

The majority of students fall into study time category 2, while very few students belong to category 4.

```python
df['Absences'].value_counts().sort_index().plot(kind='bar')
plt.title("Absence Distribution Chart")
plt.xlabel('Student Number')
plt.ylabel('Days Absent')
plt.show()
```

Most students have low absence counts.
The distribution is right-skewed, with a few students having very high absences.
There are potential outliers with unusually high absences.

```python
df['Failures'].value_counts().sort_index().plot(kind='bar')
plt.title("Failure Distribution Chart")
plt.ylabel('Student Number')
plt.xlabel('Student Failed')
plt.show()
```

The majority of students have no previous failures. The number of students decreases sharply as the number of failures increases

```python
df['Health'].value_counts().sort_values(ascending=False).plot(kind='bar')
plt.title("Health Distribution Chart")
plt.ylabel("Number of Student")
plt.xlabel("Student Category")
plt.show()
```

Most students rated their health as 4 or 5, while fewer students reported poor health (1 or 2).

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Studytime', y='G3', data=df)

plt.title("Study Time vs Final Grade")
plt.show()
```

```python
sns.scatterplot(x='Absences', y='G3', data=df)
plt.title("Identifying relationship between Final Score and Absence")
plt.ylabel("Final Score")
plt.show()
```

```python
sns.boxplot(x='Failures', y='G3', data=df)
plt.title("Identifying relationship between Final Score and Failures")
plt.ylabel("Final Score")
plt.show()
```

The median final grade decreases as the number of previous failures increases, suggesting a negative association between previous failures and academic performance.
Students with 0 previous failures generally achieve the highest grades.
Category 1 shows the largest variability in grades, indicating that while many students perform poorly, some still achieve relatively high scores.
Categories 2 and 3 have lower medians and a smaller spread, suggesting consistently lower performance.

```python
sns.boxplot(x='Health', y='G3', data=df)
plt.title("Identifying relationship between Final Score and Health")
plt.ylabel("Final Score")
plt.xlabel("Health Categories")
plt.show()
```

The box plot does not show a clear relationship between self-reported health and final grades. Students across all health categories have similar median scores and overlapping distributions, suggesting that health alone is not a strong predictor of academic performance in this dataset.

```python
sns.boxplot(x='Dalc', y='G3', data=df)
plt.title("Relationship between weekday Alcohol Consumption and Final Score")
plt.ylabel("Final Score")
plt.xlabel("Alcohol Consumption Category")
plt.show()
```

The box plot does not indicate a strong relationship between weekday alcohol consumption (Dalc) and final grades (G3). The median grades remain relatively similar across most categories, although a few low-scoring outliers are present in categories 1–4. This suggests that weekday alcohol consumption alone is not a strong predictor of academic performance in this dataset

```python
sns.boxplot(x='Walc', y='G3', data=df)
plt.title("Relationship between weekend Alcohol Consumption and Final Score")
plt.ylabel("Final Score")
plt.xlabel("Alcohol Consumption Category")
plt.show()
```

The box plot does not indicate a clear relationship between weekend alcohol consumption (Walc) and final grades (G3). The median grades remain relatively consistent across all categories, suggesting that weekend alcohol consumption alone is not a strong predictor of academic performance.
